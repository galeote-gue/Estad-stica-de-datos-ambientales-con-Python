**Plot histogram of DIC and TA**

```python
import pandas as pd

# Import the air quality dataset 
CO2Data = pd.read_csv("Terminos_lagoon_TA_DIC_2023_RawData.csv")

print(CO2Data.info())
```

```python
import matplotlib.pyplot as plt

# Select data
dic_data = CO2Data["dic_micromol_kg"]

plt.hist(dic_data, bins=8, color= "green", alpha= 0.5, histtype= 'stepfilled') #Se decidió el valor de bins en 8 con base en la regla de Sturges
plt.xlabel("Dissolved inorganic carbon (DIC) [$\mu mol/Kg$]")
plt.ylabel("Probablity")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Select data
ta_data = CO2Data["ta_micromol_kg"]

plt.hist(ta_data, bins=8, color= "red", alpha= 0.5, histtype= 'stepfilled') #Se decidió el valor de bins en 8 con base en la regla de Sturges
plt.xlabel("Total alkalinity  [$\mu mol/Kg$]")
plt.ylabel("Probablity")
plt.show()
```
